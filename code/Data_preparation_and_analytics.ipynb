```python
import pyarrow.parquet as pq
import pandas as pd
import numpy as np
from datetime import datetime
from matplotlib import pyplot as plt

data_path = '../data/'
```

```python
transactions_df = pq.read_pandas(data_path + 'transactions.parquet').to_pandas()
clients_df = pd.read_csv(data_path + 'clients.csv')
materials_df = pd.read_csv(data_path + 'materials.csv',
                           usecols=['material', 'hier_level_1', 'is_private_label', 'is_alco'])
plants_df = pd.read_csv(data_path + 'plants.csv')
```

Заметим, что в транзакциях есть как покупки так и возвраты. Разделим на два датафрейма

```python
purchase_df = transactions_df[transactions_df['sales_sum'] > 0]
return_df = transactions_df[transactions_df['sales_sum'] < 0]
del transactions_df
```

## Посмотрим на распределение дат последних покупок 

```python
first_purchase_date = purchase_df['chq_date'].min()
print('Информация о транзакциях доступна за промежуток: с {} по {} '.format(purchase_df['chq_date'].min(), purchase_df['chq_date'].max()))
```

```python
client_purchase_df = purchase_df.drop_duplicates(subset='chq_id', ignore_index=True)[['chq_date', 'client_id']]\
                        .groupby('client_id', as_index=False)\
                        .agg(lambda x : sorted(x.tolist()))

client_purchase_df = client_purchase_df.rename(columns={'chq_date' : 'purchase_dates'})
```

```python
ax = plt.axes()
plt.title('Распределение клиентов по датам последних покупок')
client_purchase_df['purchase_dates'].apply(lambda x : x[-1]).hist(figsize=(12,8), bins=50, ax=ax)
plt.show()
```

### Для того, чтобы разметить данные будем считать отточными тех клиентов, у которых последняя транзакция была совершена до 2017.05.04. Доля таких клиентов примерно 15%

```python
last_transaction_threshold = datetime(year=2017, month=5, day=4)
outflow_percentage = (client_purchase_df['purchase_dates'].apply(lambda x : x[-1]) < last_transaction_threshold).sum() / len(client_purchase_df) * 100
print('Доля отточных клиентов в нашей выборке: {}%'.format(outflow_percentage))
```

```python
client_purchase_df['is_outflow'] = client_purchase_df['purchase_dates']\
                                        .apply(lambda x : x[-1]) < last_transaction_threshold
client_purchase_df['purchase_dates'] = client_purchase_df['purchase_dates']\
                                            .apply(lambda x : list(filter(lambda i : i < last_transaction_threshold, x)))
```

### Удалим информацию о транзакциях клиентов начиная с 2017.05.01, чтобы избежать утечки информации об ответах в признаки объектов

```python
purchase_df = purchase_df[purchase_df['chq_date'] < last_transaction_threshold]
return_df = return_df[return_df['chq_date'] < last_transaction_threshold]
```

## Сформируем датасет в который будут входить признаки:

Среднее количество еды в чеке

Среднее количество не еды в чеке

is_outflow: 1 если клиент уйдет, 0 если останется

```python
import pandas as pd
import numpy as np
from datetime import datetime

class DataPrep:
    def __init__(self,
            purchase_df : pd.DataFrame,
            return_df : pd.DataFrame,
            clients_df : pd.DataFrame,
            materials_df : pd.DataFrame,
            plants_df : pd.DataFrame,
            last_transaction_threshold : datetime):
        
        self.purchase_df = purchase_df
        self.return_df = return_df
        self.clients_df = clients_df
        self.materials_df = materials_df
        self.plants_df = plants_df
        self.last_transaction_threshold = last_transaction_threshold

    def create_dataset(self):
        print('Creating client transactions df...')
        client_transactions_dataset_df = _create_client_transactions_dataset()
        print('Created!')
        
        print('Creating client info df...')
        client_info_dataset_df = _create_client_info_dataset()
        print('Created!')
        
        print('Merging...')
        dataset_df = client_transactions_dataset_df\
                    .merge(client_info_dataset_df, on='client_id', as_index=False)
        print('Merged!')
        
        return dataset_df
    
    def _create_client_transactions_dataset(self):
        # purchase processing
        print('\tMerging purchase with materials...')
        df = self.purchase_df\
            .merge(self.materials_df, on='material')
        print('\tDone!')
        print('\tMerging purchase with plants...')
        df = df.merge(self.plants_df, on='plant')
        df = pd.concat([df,
                        pd.get_dummies(df['hier_level_1'], prefix='product')],
                        axis=1)\
                        .drop(labels=['chq_position', 'material', 'sales_count', 'hier_level_1'], axis=1)
        print('\tDone!')
        
        print('\tAggregating purchase by checks...')
        chq_agg_func = {
            'client_id' : lambda x : x.iloc[0],
            'plant_type' : lambda x : x.iloc[0],
            'plant' : lambda x : x.iloc[0],

            'sales_sum' : sum,
            'is_promo' : sum,
            'is_private_label' : sum,
            'is_alco' : sum,
            'product_FOOD' : sum,
            'product_NONFOOD' : sum
        }
        df = df.groupby('chq_id', as_index=False).agg(chq_agg_func)
        print('\tDone!')
        
        print('\tAggregating purchase by clients...')
        client_agg_func = {
            'plant_type' : lambda x : x.value_counts().idxmax(),
            'plant' : lambda x : len(x.unique()),

            'sales_sum' : lambda x : x.mean(),
            'is_promo' : lambda x : x.mean(),
            'is_private_label' : lambda x : x.mean(),
            'is_alco' : lambda x : x.mean(),
            'product_FOOD' : lambda x : x.mean(),
            'product_NONFOOD' : lambda x : x.mean()
        }
        df = df.groupby('client_id', as_index=False).agg(client_agg_func)
        print('\tDone!')
        
        # return processing
        print('\tMerging returns with materials...')
        df1 = self.return_df\
            .merge(self.materials_df, on='material')
        print('\tDone!')
        
        print('\tAggregating returns by checks...')
        df1 = df1[['chq_id', 'client_id', 'sales_sum']]\
            .groupby('chq_id', as_index=False)\
            .agg({
            'client_id' : lambda x : x.iloc[0],
            'sales_sum' : sum
        })
        print('\tDone!')
        
        print('\tAggregating returns by clients...')
        df1.groupby('client_id', as_index=False)\
            .agg({
            'sales_sum' : lambda x : np.abs(x.mean())
        })
        print('\tDone!')
        
        df1 = df1.rename(columns={'sales_sum' : 'return_sum'})
        
        df = df.rename(columns={
            'plant_type' : 'most_freq_plant_type',
            'plant' : 'diff_plants_amount',
            'sales_sum' : 'check_sum',
            'is_promo' : 'promo_amount',
            'is_private_label' : 'priv_lbl_amount',
            'is_alco' : 'alco_amount',
            'product_FOOD' : 'food_amount',
            'product_NONFOOD' : 'nonfood_amount'
        })
        
        # concate ???
        # merging dataframes
        df = df.merge(df1, on='client_id')
        
        return df
```
